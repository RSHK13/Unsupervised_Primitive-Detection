```python
def calculate_entropy(text, word, flag):
    word_count = Counter()
    conditional_count = Counter()
    entropy = 0

    for i in range(len(text) - len(word) + 1):
        if text[i:i + len(word)] == word:
            word_count[word] += 1
            if i + len(word) < len(text):
                next_letter = text[i + len(word)]
                conditional_count[next_letter] += 1

    p_word = word_count[word] / (len(text) - len(word) + 1)
    ab = word_count[word]
    if ab == 1 :
      entropy = 0
    else:
      for letter, count in conditional_count.items():
        p_letter_given_word = count / word_count[word]
        entropy -= p_letter_given_word * math.log2(p_letter_given_word)
    if flag == 0:
      return entropy
    else:
      return entropy,ab
```

```python
from collections import Counter
from collections import defaultdict
import math

def calculate_expected_entropy(text,n):
    bigram_frequencies = defaultdict(int)
    sum = 0
    sum2 = 0

    # Iterate over the text to count the bigram frequencies
    for i in range(len(text)-(n-1)):
        bigram = text[i:i+n]
        bigram_frequencies[bigram] += 1

    for word,freq in bigram_frequencies.items():
        entropy = calculate_entropy(user_text,word,0)
        if entropy >0 :
            sum = sum + freq*entropy
            sum2 = sum2 + freq

    average_entropy = sum / sum2

    return average_entropy

# text = user_text
# average_entropy = calculate_expected_entropy(text,13)
# print(f"Average bigram entropy: {average_entropy}")
```

```python
def find_words(text, expected_entropy):
    words = []

    i = 0
    while i < len(text):
        max_entropy = 0.0
        word_identified = False

        for j in range(i+1, min(i + 14, len(text))):
            word = text[i:j+1]
            word_length = len(word)
            prev_word = text[i:j]
            next_word = text[i:min(j+2,len(text)-1)]
            entropy = calculate_entropy(text, word,0)/pow(expected_entropy[len(word)-1],2)
            prev_entropy = calculate_entropy(text, prev_word,0)/pow(expected_entropy[len(prev_word)-1],2)
            next_entropy = calculate_entropy(text, next_word,0)/pow(expected_entropy[len(next_word)-1],2)

            if entropy > prev_entropy and entropy >= next_entropy :
                max_entropy_index = j
                word_identified = True
                break
            # if len(word) == 2 and entropy*1.414 > prev_entropy and entropy*1.414 > next_entropy*1.732:
            #     max_entropy_index = j
            #     word_identified = True
            #     break

        if word_identified:
            words.append(text[i:max_entropy_index+1])
            i = max_entropy_index + 1
        else:
            words.append(text[i])
            i += 1


    return words
```

```python
def calculate_reconstruction_loss(corpus, identified_words):
    skipped_characters = 0
    identified_text = ''.join(identified_words)
    i = 0
    j = 0

    while i < len(corpus) and j < len(identified_text):
        if corpus[i] == identified_text[j]:
            i += 1
            j += 1
        else:
            skipped_characters += 1
            i += 1

    skipped_characters += len(corpus) - i

    reconstruction_loss = skipped_characters / len(corpus)
    return reconstruction_loss

reconstruction_loss = calculate_reconstruction_loss(user_text, identified_words)
print("Reconstruction loss:", reconstruction_loss)
```

```python
def calculate_Length_parameter(identified_words):
    freq = {}
    length_param = 0
    lp = 0
    for word in identified_words:
        freq[word] = freq.get(word, 0) + 1
        length_param += freq[word]
        word_length = len(word)
        lp += freq[word]*word_length

    length_param = length_param/lp
    return length_param

length_inv = calculate_Length_parameter(identified_words)
print("Lenght parameter:",length_inv)
```

```python
def optimize_words(identified_words, text, expected_entropy):

        i = 0
        while i < len(identified_words) :


          word = identified_words[i]
          entropy,freq = calculate_entropy(text, word,1)
          if len(word) > 23:
            entropy = entropy/pow(expected_entropy[23],0.5)
          else:
            entropy = entropy /pow(expected_entropy[len(word)-1],0.5)
          if entropy == 0 and freq > 1:
            i = i + 1
            continue
          else:
            index  = 0
            best_MI = -1
            word_length = len(word)

            for split_point in range(1, word_length):

                part_A = word[:split_point]
                part_B = word[split_point:]


                entA,freqA = calculate_entropy(text, part_A,1)
                if len(part_A) > 23:
                    entA = entA/pow(expected_entropy[23],0.5)
                else:
                    entA = entA /pow(expected_entropy[len(part_A)-1],0.5)
                entB,freqB = calculate_entropy(text, part_B,1)
                if len(part_B) > 23:
                      entB = entB/pow(expected_entropy[23],0.5)
                else:
                      entB = entB /pow(expected_entropy[len(part_B)-1],0.5)
                if entA > entropy and entB > entropy:
                  if freqA > freq and freqB > freq:
                    if entA + entB - entropy > best_MI and entA >0 and entB>0:
                      index = split_point
                      best_MI = entA + entB - entropy
                      bA = part_A
                      bB = part_B

            if index == 0:
                i = i + 1
                continue
            else :
                identified_words = replace_index(identified_words, i, bA)
                identified_words = add_index(identified_words, i+1, bB)
                i = i + 2
        return identified_words
```

```python
def replace_index(arr, index, new_string):
    if index < 0 or index >= len(arr):
        print("Index out of range!")
        return arr

    new_arr = arr.copy()  # Create a copy of the original array to avoid modifying it directly
    new_arr[index] = new_string
    return new_arr
```

```python
def remove_index(arr, index):
    if index < 0 or index >= len(arr):
        print("Index out of range!")
        return arr

    new_arr = arr.copy()  # Create a copy of the original array to avoid modifying it directly
    del new_arr[index]
    return new_arr
```

```python
def add_index(arr, index, value):
    if index < 0 or index > len(arr):
        print("Index out of range!")
        return arr

    new_arr = arr.copy()  # Create a copy of the original array to avoid modifying it directly
    new_arr.insert(index, value)
    return new_arr
```

```python
def add_consequent_words(text, identified_words, expected_entropy):
    i = 0
    new_words = []
    while i < len(identified_words) - 2:
        new_word = ""
        j = i
        word = identified_words[i]
        best_index = i
        best_diff = 0
        diff = 0
        # if len(word) > 23:
        #     current_entropy = pow(expected_entropy[23],1.75)
        # else:
        #     current_entropy = pow(expected_entropy[len(word)-1],1.75)
        # ent = calculate_entropy(text, word, 0) / current_entropy

        while j < min(i + 7, len(identified_words)):
            prev_word = new_word
            new_word += identified_words[j]
            if j == len(identified_words)-1:
                next_word = new_word
            else:
                next_word = new_word + identified_words[j+1]
            if len(new_word) < 15:
                if prev_word == "":
                    prev_entropy = 0
                else:
                    prev_entropy = calculate_entropy(text, prev_word, 0) / pow(expected_entropy[len(prev_word)-1],1.5)
                new_entropy = calculate_entropy(text, new_word, 0)
                new_entropy = new_entropy / pow(expected_entropy[len(new_word)-1],1.5)
                if len(next_word) > 20:
                  next_entropy = 0
                else:
                  next_entropy = calculate_entropy(text, next_word, 0) / pow(expected_entropy[len(next_word)-1],1.5)

                if new_entropy > prev_entropy and new_entropy > next_entropy:
                    diff = new_entropy - next_entropy
                else:
                    diff = 0
                if diff >= best_diff:
                    best_diff = diff
                    best_index = j
                    word = new_word

            j += 1

        new_words.append(word)
        i = best_index + 1

    return new_words
```
